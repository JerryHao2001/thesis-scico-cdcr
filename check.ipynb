```python
# Cell 1: imports
from datasets import load_dataset
import pandas as pd
import numpy as np
from collections import Counter
import math
import matplotlib.pyplot as plt

plt.style.use("default")  # keep it simple
```

```python
# Cell 2: load SciCO train split
scico = load_dataset("allenai/scico")["train"]
len(scico), scico[0].keys()
```

```python
# Cell 3: helper to compute stats for one topic
def topic_stats(row):
    topic_id = row["id"]
    mentions = row["mentions"]  # list of [doc_id, start, end, cluster_id]
    tokens_per_doc = row["tokens"]  # list of [tokens]
    
    n_mentions = len(mentions)
    n_docs = len(tokens_per_doc)
    n_tokens = sum(len(doc_tokens) for doc_tokens in tokens_per_doc)
    
    # cluster sizes
    cluster_ids = [m[3] for m in mentions]
    cluster_counts = Counter(cluster_ids)
    cluster_sizes = list(cluster_counts.values())
    n_clusters = len(cluster_counts)
    max_cluster_size = max(cluster_sizes) if cluster_sizes else 0
    min_cluster_size = min(cluster_sizes) if cluster_sizes else 0
    
    # pairwise stats over mentions
    if n_mentions >= 2:
        total_pairs = n_mentions * (n_mentions - 1) // 2
    else:
        total_pairs = 0
        
    pos_pairs = sum(c * (c - 1) // 2 for c in cluster_sizes)
    pos_rate = pos_pairs / total_pairs if total_pairs > 0 else 0.0
    
    # mentions / tokens density
    mentions_per_100_tokens = (n_mentions / n_tokens * 100) if n_tokens > 0 else 0.0
    
    return {
        "id": topic_id,
        "n_docs": n_docs,
        "n_tokens": n_tokens,
        "n_mentions": n_mentions,
        "n_clusters": n_clusters,
        "max_cluster_size": max_cluster_size,
        "min_cluster_size": min_cluster_size,
        "mentions_per_100_tokens": mentions_per_100_tokens,
        "total_pairs": total_pairs,
        "pos_pairs": pos_pairs,
        "pos_rate": pos_rate,
    }
```

```python
# Cell 4: build a DataFrame of topic-level stats
rows = [topic_stats(scico[i]) for i in range(len(scico))]
df = pd.DataFrame(rows)
df.head()
```

```python
# Cell 5: summary statistics
df.describe(percentiles=[0.1, 0.25, 0.5, 0.75, 0.9]).T
```

```python
# Cell 6: n_mentions per topic
plt.figure(figsize=(6,4))
plt.hist(df["n_mentions"], bins=40)
plt.yscale("log")
plt.xlabel("Number of mentions per topic")
plt.ylabel("Count (log scale)")
plt.title("SciCO train: mentions per topic")
plt.show()
```

```python
# Cell 7: n_clusters per topic
plt.figure(figsize=(6,4))
plt.hist(df["n_clusters"], bins=40)
plt.yscale("log")
plt.xlabel("Number of clusters per topic")
plt.ylabel("Count (log scale)")
plt.title("SciCO train: clusters per topic")
plt.show()
```

```python
# Cell 8: positive pair fraction per topic
plt.figure(figsize=(6,4))
plt.hist(df["pos_rate"], bins=40)
plt.xlabel("Positive pair fraction within-topic")
plt.ylabel("Count")
plt.title("SciCO train: positive pair fraction per topic")
plt.show()
```

```python
# Cell 9: mentions density
plt.figure(figsize=(6,4))
plt.hist(df["mentions_per_100_tokens"], bins=40)
plt.xlabel("Mentions per 100 tokens")
plt.ylabel("Count")
plt.title("SciCO train: mention density")
plt.show()
```

```python
# Cell 10: collect all cluster sizes
all_cluster_sizes = []

for i in range(len(scico)):
    mentions = scico[i]["mentions"]
    cluster_ids = [m[3] for m in mentions]
    cluster_counts = Counter(cluster_ids)
    all_cluster_sizes.extend(cluster_counts.values())

sizes = np.array(all_cluster_sizes)
print("Total clusters:", len(sizes))
print("Cluster size stats:")
print("  min:", sizes.min(), "max:", sizes.max())
print("  mean:", sizes.mean())
print("  median:", np.median(sizes))
print("  90th pct:", np.percentile(sizes, 90))
```

```python
# Cell 11: histogram of cluster sizes
plt.figure(figsize=(6,4))
plt.hist(sizes, bins=range(1, sizes.max()+2))
plt.yscale("log")
plt.xlabel("Cluster size")
plt.ylabel("Count (log scale)")
plt.title("SciCO train: cluster size distribution")
plt.show()
```
